```python
from dotenv import load_dotenv
import os
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import time
import numpy as np

load_dotenv(r"C:\Users\leonh\source\repos\DataVisualizationLabs\FFLogs_ID_Secret.env")

V1_KEY = os.getenv("FFLOGS_V1_KEY")
```

```python
# mapping FFLogs (class, spec) → FFXIV job names
job_map = {
    1: "Paladin",
    2: "Warrior",
    3: "Dark Knight",
    4: "Gunbreaker",

    5: "White Mage",
    6: "Scholar",
    7: "Astrologian",
    26: "Sage",

    8: "Monk",
    9: "Dragoon",
    10: "Ninja",
    11: "Samurai",
    28: "Reaper",

    12: "Bard",
    13: "Machinist",
    14: "Dancer",

    15: "Black Mage",
    16: "Summoner",
    17: "Red Mage",
}
```

```python
# finds the encounter ids

url = f"https://www.fflogs.com/v1/zones?api_key={V1_KEY}"
zones = requests.get(url).json()

#use for enc codes
#for zone in zones:
 #   print("\nZONE:", zone['name'])
  #  for enc in zone.get('encounters', []):
   #     print("  ", enc['id'], "-", enc['name'])
```

```python
# list of enc
encounters = {
    "P9S": 89,
    "P10S": 90,
    "P11S": 91,
    "P12S": 92,
    "UCOB": 1047,
    "UWU": 1048,
    "DSR": 1061,
    "TOP": 1062
}
```

```python
#get all job of all enc
all_rows = []

for fight_name, enc_id in encounters.items():
    for spec_id, job_name in job_map.items():

        url = f"https://www.fflogs.com/v1/rankings/encounter/{enc_id}?metric=rdps&spec={spec_id}&api_key={V1_KEY}"
        data = requests.get(url).json()

        # Skip errors
           # Only skip if DATA DOES NOT contain rankings
        if not isinstance(data, dict) or "rankings" not in data:
            continue


        for row in data.get("rankings", []):
            row["job"] = job_name
            row["fight"] = fight_name
            all_rows.append(row)

        time.sleep(0.25)

df_all_jobs = pd.DataFrame(all_rows)
df_all_jobs.head()
```

```python
#cleans unneeds damage types
df_clean = df_all_jobs[df_all_jobs["job"].notna()]
df_clean = df_clean[~df_clean["job"].isin(["LimitBreak", "Unknown", "???"])]
```

```python
# adds fight tiers
df_clean["tier"] = df_clean["fight"].apply(
    lambda f: "Savage" if f.startswith("P") else "Ultimate"
)
```

```python
pop_all = pd.read_csv(r"D:\School\SIM3032-O\WeekFour\ClassDistChart.csv")
pop_end = pd.read_csv(r"D:\School\SIM3032-O\WeekFour\EndGameClassDistChart.csv")
```

```python
#maps census to job
census_to_job = {
    "Gladiator": "Paladin",
    "Marauder": "Warrior",
    "Dark Knight": "Dark Knight",
    "Gunbreaker": "Gunbreaker",

    "Conjurer": "White Mage",
    "Scholar": "Scholar",
    "Astrologian": "Astrologian",
    "Sage": "Sage",

    "Lancer": "Dragoon",
    "Pugilist": "Monk",
    "Rogue": "Ninja",
    "Samurai": "Samurai",
    "Reaper": "Reaper",

    "Archer": "Bard",
    "Machinist": "Machinist",
    "Dancer": "Dancer",

    "Thaumaturge": "Black Mage",
    "Arcanist": "Summoner",
    "Red Mage": "Red Mage"
}
```

```python
# Copy endgame census
pop_end_mapped = pop_end.copy()

# Convert class → job
pop_end_mapped["job"] = pop_end_mapped["Category"].map(census_to_job)

# Remove unmapped rows
pop_end_mapped = pop_end_mapped.dropna(subset=["job"])

# Rename for merge
pop_end_mapped = pop_end_mapped.rename(columns={"Active Characters": "active_players"})

# Compute popularity percent
total_active = pop_end_mapped["active_players"].sum()
pop_end_mapped["popularity_percent"] = (
    pop_end_mapped["active_players"] / total_active * 100
)

display(pop_end_mapped)
```

```python
# clean final 
df_final = df_clean.merge(
    pop_end_mapped[["job", "popularity_percent"]],
    on="job",
    how="left"
)
```

# VISUALIZATION

## Job Popularity vs rDPS (Scatterplot)

```python
plt.figure(figsize=(14,8))

# Scatter plot
sns.scatterplot(
    data=df_final,
    x="popularity_percent",
    y="rDPS",
    hue="job",
    alpha=0.7,
    s=80
)

plt.title("FFXIV Job Popularity vs rDPS")
plt.xlabel("Job Popularity (%)")
plt.ylabel("rDPS")

# Legend
plt.legend(
    title="Job",
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

# Extend y-axis downward to make space
ymin, ymax = plt.ylim()
plt.ylim(ymin - 2000, ymax)   

# --- Add markers + job labels ---
jobs = df_final["job"].unique()

for job in jobs:
    x_val = df_final[df_final["job"] == job]["popularity_percent"].iloc[0]

    # downward marker on 0 line
    plt.plot(x_val, ymin - 50, marker="v", color="black")

    # rotated job label below chart
    plt.text(
        x_val,
        ymin - 100,      # stays inside the extended y-range
        job,
        ha='center',
        va='top',
        fontsize=8,
        rotation=90
    )

plt.tight_layout()
plt.show()
```

### What this visualization shows

This scatterplot compares each job’s overall player popularity with its average raid DPS (rDPS) across high-level content (Savage + Ultimate logs). Each job appears as a colored point representing:

- **X-axis:** How common the job is among active players (popularity %)  
- **Y-axis:** The job’s performance in raid settings (average rDPS)  
- **Color:** Identifies which job each point represents  

This visualization provides a clear overview of how job performance and player preference align (or don’t align) within FFXIV’s endgame community.

---

### What you can see clearly

Some jobs with high popularity—such as **Summoner** and **Samurai**—sit near the middle or upper end of the DPS spectrum. In contrast, jobs like **Astrologian** and **Monk** show **lower overall popularity** despite **relatively high rDPS performance**. This contrast suggests that factors such as job difficulty, mechanical complexity, or a steep skill ceiling may influence whether players choose a job, even when the job performs well in high-end content.

When examining **Ninja, Dancer, Bard, and Red Mage**, these jobs fall on the **lower end of rDPS**, yet their popularity is spread widely across the chart. This indicates that player preference is not strictly tied to raw damage output. Qualities such as utility, mobility, support value, rotation feel, and accessibility likely explain why these jobs remain popular despite dealing less damage than higher-performing roles.

---

### Conclusion

The scatterplot demonstrates that **FFXIV job popularity does not strongly correlate with raw raid performance**. Many players choose jobs for reasons beyond DPS, including aesthetics, role fantasy, utility, personal comfort, and approachability. As a result, job popularity reflects a balance of mechanical strength and player preference, rather than being driven solely by damage potential.

### Conclusion

The rDPS boxplot reveals not only which jobs tend to perform the best on average, but also how consistent their performance is across the playerbase. Jobs with tight interquartile ranges tend to have more predictable outputs, while those with wider distributions may be more sensitive to player skill, encounter mechanics, or rotation difficulty. ## Savage vs Ultimate rDPS Comparison

```python
plt.figure(figsize=(16,8))

sns.boxplot(
    data=df_final,
    x="fight",       # each fight gets its own box
    y="rDPS",
    palette="Set3"
)

plt.title("rDPS Distribution by Individual Encounter")
plt.xlabel("Encounter")
plt.ylabel("rDPS")
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.show()
```

### What this visualization shows:
This boxplot compares the rDPS output for each individual raid encounter, across both
Savage and Ultimate fights. Each box represents the spread of DPS performance for all
jobs within a specific encounter, allowing you to see how tightly (or widely) players
perform in each fight.

### What you can see clearly:
- Savage fights (P9S–P12S) tend to show **higher rDPS values overall**, because they are
  not item-level synced and players revisit them with higher gear, stronger weapons,
  and access to later-game optimization.

- Ultimate fights (UCOB, UWU, DSR, TOP) display **consistently lower rDPS ranges**.  
  However, this is **not** because they are older — it is due to their **item-level sync**.
  Ultimates lock players to a fixed gear cap, meaning rDPS cannot be inflated with
  new expansions or better gear.

### Why this matters:
Because Ultimates are item-level synced, their DPS ceiling stays the same permanently.
Any improvements in their rDPS over time come only from:
- increased player skill,
- community-optimized routes and openers,
- tighter execution and buff alignment,
- and refined strategies developed over years.

This means Ultimate fights naturally appear lower on the rDPS scale, but their numbers
are a more direct reflection of **player mastery** rather than gear progression.

### Conclusion:
Savage encounters demonstrate how gear progression boosts rDPS over time, while
Ultimate encounters highlight consistent performance bounds due to item-level sync.
Together, these plots show the contrast between “gear-influenced” DPS growth in Savage
fights and “skill-only” performance in Ultimate content.


 ## Average vs Highest rDPS by Job (Bar Chart)

```python
# Compute avg and max rDPS per job
job_compare = df_final.groupby("job").agg(
    avg_rDPS=("rDPS", "mean"),
    max_rDPS=("rDPS", "max")
).reset_index()

# Sort jobs by avg rDPS descending
job_compare = job_compare.sort_values("avg_rDPS", ascending=False)

import numpy as np

plt.figure(figsize=(18,8))

x = np.arange(len(job_compare))
width = 0.4

# Two bar sets: avg and max
plt.bar(x - width/2, job_compare["avg_rDPS"], width, label="Average rDPS", color="skyblue")
plt.bar(x + width/2, job_compare["max_rDPS"], width, label="Highest rDPS", color="orange")

# Add labels above bars
for i, v in enumerate(job_compare["avg_rDPS"]):
    plt.text(i - width/2, v + 30, f"{v:.0f}", ha='center', fontsize=8)

for i, v in enumerate(job_compare["max_rDPS"]):
    plt.text(i + width/2, v + 30, f"{v:.0f}", ha='center', fontsize=8)

plt.xticks(x, job_compare["job"], rotation=45, ha="right")
plt.ylabel("rDPS")
plt.title("Average vs Highest rDPS by Job")
plt.legend()

plt.tight_layout()
plt.show()
```

## What this visualization shows:
This bar chart compares the average rDPS and the highest recorded rDPS for every job in the dataset. Each job has two bars placed side by side:
- One bar represents the average rDPS across all logs.
- The second bar shows the single highest rDPS achieved by any player for that job.

This allows us to see not only how each job typically performs, but also how far top players can push the job when fully optimized.

## What you can see clearly:
- Jobs like Black Mage, Samurai, and Monk have some of the largest gaps between average and maximum rDPS. This indicates very high skill ceilings. Skilled players can push these jobs much further than the average player.
  
- Jobs with smaller gaps, such as Bard, Dancer, Red Mage, and Ninja, show more consistent performance across the playerbase. These jobs tend to have simpler or more supportive kits, which leads to less variation between average and top-end players.

- Several non-DPS jobs, including White Mage, Dark Knight, and Astrologian, appear surprisingly high on the maximum rDPS list. This happens because specific encounter windows allow them to output huge burst phases, even though these bursts do not reflect their typical raid role or sustained output.

## Why this matters:
This chart helps highlight:
- Which jobs reward mastery with significantly higher performance.
- Which jobs maintain stable output regardless of player skill.
- How job complexity, rotation strictness, and encounter design influence potential damage.

It also reinforces that average performance and peak potential are two separate aspects when evaluating job balance.

## Conclusion:
The difference between average and highest rDPS clearly shows each job's skill ceiling. High skill ceiling jobs rise dramatically under expert play, while support-oriented or forgiving jobs maintain steadier performance. This visualization reveals how much optimization influences job output and why certain jobs dominate speedrunning or top percentile logs, even if they are not the most popular choices among the general playerbase.


## Personal DPS vs Raid DPS by Job

```python

# Prepare data
job_dps = df_clean.groupby("job")[["nDPS", "rDPS"]].mean().reset_index()
job_dps = job_dps.sort_values("nDPS", ascending=False)

jobs = job_dps["job"]
nDPS = job_dps["nDPS"]
rDPS = job_dps["rDPS"]

x = np.arange(len(jobs))
width = 0.38

plt.figure(figsize=(16, 9))

# Make sure the plot has more space above the bars
plt.ylim(0, max(max(nDPS), max(rDPS)) * 1.15)

bars1 = plt.bar(x - width/2, nDPS, width, label="Personal DPS (nDPS)", color="skyblue")
bars2 = plt.bar(x + width/2, rDPS, width, label="Raid DPS (rDPS)", color="orange", alpha=0.8)

# Label both bars
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + (height * 0.015),
             f"{height:.0f}", ha='center', va='bottom', rotation=90, fontsize=9)

for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + (height * 0.015),
             f"{height:.0f}", ha='center', va='bottom', rotation=90, fontsize=9, color="darkred")

plt.xticks(x, jobs, rotation=75)
plt.ylabel("DPS")
plt.title("nDPS vs rDPS by Job (Side-by-Side Bars, Expanded Top Margin)")
plt.legend()
plt.tight_layout()
plt.show()
```

## What this visualization shows:
This bar chart compares each job's personal DPS (nDPS) with its raid-adjusted DPS contribution (rDPS). Each job has two bars side by side:

- nDPS represents the player's raw personal damage output with no raid buffs applied.
- rDPS represents the raid-adjusted value, which includes the extra DPS gained from party buffs.

This comparison highlights how much each job relies on personal damage versus raid synergy.

## What you can see clearly:
- Jobs with high nDPS and only a small increase in rDPS, such as Samurai and Black Mage, generate most of their output through personal rotation rather than buff scaling.

- Jobs with lower nDPS but a much larger jump in rDPS, such as Bard, Dancer, Ninja, and Red Mage, benefit significantly from raid buffs. These jobs often provide utility, support tools, or buff windows that enhance total party damage.

- Tanks and healers appear lower on the chart overall. Their rDPS is only slightly higher than their nDPS, showing that they gain less from raid buffs because their kits are not focused on maximizing personal damage.

- The size of the gap between nDPS and rDPS helps reveal which jobs scale heavily with optimized teams and which perform consistently regardless of composition.

## Why this matters:
Comparing nDPS and rDPS provides insight into how a job functions in coordinated group content. It shows:

- Which jobs rely primarily on personal damage.  
- Which jobs gain more value from synergy and buff windows.  
- How job design choices influence overall raid contribution.

This makes the chart useful for evaluating job scaling, understanding synergy, and identifying which jobs thrive in optimized groups compared to casual settings.

## Conclusion:
This visualization clearly separates jobs that excel through raw personal damage from jobs that shine through party synergy. High nDPS jobs like Black Mage and Samurai remain strong on their own, while support-leaning jobs such as Bard, Dancer, Red Mage, and Ninja scale dramatically with raid buffs. This emphasizes how FFXIV balances its roster between self-sufficient damage dealers and utility-focused jobs that boost overall party output.

## DPS Metric Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_final[["aDPS","rDPS","nDPS","cDPS","pDPS"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between DPS Metrics")
plt.show()
```

## What this visualization shows:
This heatmap displays the correlation between the five major DPS metrics used in FFLogs: aDPS, rDPS, nDPS, cDPS, and pDPS. Each cell shows how strongly two metrics move together using a correlation value between 0 and 1, where higher numbers indicate stronger relationships.

- aDPS = adjusted DPS (removes padding)
- rDPS = raid DPS (includes raid buff contribution)
- nDPS = personal raw DPS
- cDPS = DPS contributed by buffs to others
- pDPS = pure DPS calculation similar to aDPS

## What you can see clearly:
- rDPS has a strong correlation with aDPS and nDPS. This shows that raid DPS still depends heavily on consistent personal performance, even when raid buffs are factored in.

- aDPS and nDPS are very closely correlated, meaning most personal DPS is consistent whether you look at adjusted or raw values.

- cDPS shows weaker correlations compared to the other metrics. This makes sense because cDPS measures how much a job increases *other players’* damage, which varies by job role and raid buffs.

- pDPS correlates strongly with both aDPS and nDPS, indicating that pure DPS aligns closely with personal output.

## Why this matters:
Understanding these correlations helps explain how FFLogs measures player performance. It shows which metrics rely on player skill alone and which metrics depend on party synergy or buff alignment.

This is useful for:
- Comparing job roles fairly  
- Interpreting log performance beyond raw DPS  
- Seeing how support jobs differ from pure DPS jobs  

## Conclusion:
The heatmap reveals that most DPS metrics are strongly connected, especially those tied to personal performance. cDPS stands out as the least correlated, highlighting its unique role in tracking party buff contribution. Overall, the heatmap helps clarify how each DPS metric reflects different aspects of player and party performance.


# Dataset Description

For this lab, I combined two sources of data related to Final Fantasy XIV:

### 1. FFLogs High-End Raid Performance Data
This dataset was collected using the FFLogs public API. It includes job performance statistics for eight high-end encounters:
- P9S, P10S, P11S, P12S (Savage tier)
- UCOB, UWU, DSR, TOP (Ultimate tier)

For each job and encounter, the API returned:
- nDPS: Personal raw DPS
- rDPS: Raid-adjusted DPS including buff contribution
- aDPS: Adjusted DPS with padding removed
- cDPS: DPS contributed to others through buffs
- pDPS: Pure DPS calculation
- Job name and encounter tier (Savage or Ultimate)
- Patch number and server region
- Player identifiers (not used in analysis)

This produced a dataset with over 13,000 rows of performance logs across all jobs.

### 2. FFXIV Job Popularity Census
I also used the player census data from https://ffxivcensus.com/2024-05, processed into:
- ClassDistChart.csv: Total number of active characters per job
- EndGameClassDistChart.csv: Number of active endgame characters per job

From these files, I calculated:
- Popularity percent for each job
- Position of each job relative to overall player activity

### Final Combined Dataset
After cleaning and merging, the final dataset contained:
- Job name
- Encounter name and tier
- Personal DPS metrics (nDPS, aDPS, pDPS)
- Raid DPS metrics (rDPS, cDPS)
- Popularity percentage
- Region and server identifiers

This combined dataset allowed for a full comparison of:
- Player job choice
- Personal vs raid performance
- DPS differences across encounters
- Performance trends between Savage and Ultimate content

The comparison between nDPS and rDPS highlighted how some jobs gain significant value from party buffs, while others rely more heavily on strong personal rotation execution.

The encounter-based comparisons also showed how fight design affects performance. Savage encounters allow higher DPS ceilings, while Ultimate encounters remain capped by stricter item level syncing and more punishing mechanics. The average versus maximum DPS chart also illustrated which jobs have high skill ceilings, and which jobs deliver more consistent performance across the playerbase.

It is important to note that the dataset does not include all jobs in the game. For example, Sage and Reaper were excluded from this analysis because the FFLogs dataset used did not contain their entries at the time of collection. Including additional data from later patches, more recent raid tiers, and full job release timelines would allow for a more complete comparison across all roles. Future work could also examine how job performance evolves across expansions, how balance patches shift the meta, and how job release dates influence long-term popularity trends.

Overall, the findings indicate that FFXIV job balance is designed around both personal output and party synergy, creating a system where player preference, job identity, encounter mechanics, and group coordination all interact. Performance cannot be judged purely on damage alone. The combined visualizations offer a detailed view of this ecosystem and demonstrate how multiple factors shape both job performance and job choice in high-end gameplay.
